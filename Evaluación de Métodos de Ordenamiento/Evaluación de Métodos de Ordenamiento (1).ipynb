```python
import random
import time
import matplotlib.pyplot as plt

# -------------------------------
# Algoritmo 1: Burbuja
# -------------------------------
def burbuja(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# -------------------------------
# Algoritmo 2: Quicksort
# -------------------------------
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivote = arr[0]
        menores = [x for x in arr[1:] if x <= pivote]
        mayores = [x for x in arr[1:] if x > pivote]
        return quicksort(menores) + [pivote] + quicksort(mayores)

# -------------------------------
# Función para probar ambos algoritmos
# -------------------------------
def evaluar_algoritmos(tamaño_lista):
    lista_original = random.sample(range(1, tamaño_lista * 10), tamaño_lista)

    # Evaluar Burbuja
    inicio_burbuja = time.time()
    burbuja(lista_original.copy())
    fin_burbuja = time.time()
    tiempo_burbuja = fin_burbuja - inicio_burbuja

    # Evaluar Quicksort
    inicio_quick = time.time()
    quicksort(lista_original.copy())
    fin_quick = time.time()
    tiempo_quick = fin_quick - inicio_quick

    return tiempo_burbuja, tiempo_quick

# -------------------------------
# Comparar para diferentes tamaños
# -------------------------------
tamaños = [100, 200, 500, 1000]
tiempos_burbuja = []
tiempos_quick = []

for t in tamaños:
    tb, tq = evaluar_algoritmos(t)
    tiempos_burbuja.append(tb)
    tiempos_quick.append(tq)

# -------------------------------
# Visualización
# -------------------------------
plt.figure(figsize=(10, 6))
plt.plot(tamaños, tiempos_burbuja, label='Burbuja', marker='o')
plt.plot(tamaños, tiempos_quick, label='Quicksort', marker='s')
plt.title('Comparación de Tiempos de Ejecución')
plt.xlabel('Tamaño de la Lista')
plt.ylabel('Tiempo (segundos)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
